### Переименование столбцов и замена типов данных

```python
# Приводим имена столбцов к нижнему регистру

data.columns= data.columns.str.lower()

# Проверяем
print(data.columns) # Все гуд
```

```python
# Типы данных не везде верные. Меняем year_of_release на целочисленный тип данных Int, чтобы можно было выполнять математические операции.
# Колонки genre и rating в категориальный тит данных. В принципе это необязательно, но так будет по фэншую. user_score перевенем во float.

#pd.DatetimeIndex(data.year_of_release)

#Приведем тип данных к Int
data['year_of_release'] = data['year_of_release'].astype('Int32')

#Приведем тип данных к int
#data['critic_score'] = data['critic_score'].astype('Int64')

#Приведем тип данных к категориальному
data['genre'] = data['genre'].astype('category')
print(data['genre'].dtype)

data['rating'] = data['rating'].astype('category')
#data['platform'] = data['platform'].astype('category')
#print(data['platform'].dtype)

# Проверяем
print(data.info()) # все гуд
```

### Обработка пропусков

```python
# Посмотрим сколько всего пропусков, %

display(pd.DataFrame(round(data.isna().mean()*100, )).style.background_gradient('coolwarm'))
```

Все гуд, кроме последних 3ех колонок. Слишком много пропусков

```python
# Посмотрим на количество всех уникальных значений

for i in data.columns :
    print(f'Уникальных значений колонки {i}: ', len(data[i].unique()))

#Проверим явные дубли
print("Явных дублей в таблице:", data.duplicated().sum()) # Явных дублей в таблице нет
```

```python
# В колонке name есть немало дублей. Похоже, что это связано с тем, что одна и таже игра есть на нескольких платформах. 
# Будем это использовать при заполнении пропусков.

# В колоне year_of_release 269 пропусков, удаляем эти строки т.к. они на не понадобятся
data = data.query('year_of_release.isna() == False').reset_index(drop=True)
```

```python
#Заполняем пустые значения колонки rating значениями тех же игр, но на других платформах.
# Не везде одна игра на всех платформах имеет одно и тоже значение, но мы возьмем первое, если есть хоть одно в записях

print("Значения рейтингов до: ", data.groupby('rating')['rating'].count())    
for i in data.name.unique():
    if data.loc[(data['name'] == i) & (~data['rating'].isna()), 'rating'].value_counts().max() > 0:
        data.loc[(data['name'] == i) & (data['rating'].isna()), 'rating'] = \
        data.loc[(data['name'] == i) & (~data['rating'].isna()), 'rating'].value_counts().index[0]


print("Значения рейтингов после: ", data.groupby('rating')['rating'].count())
data.info()
```

```python
# Посмотрим сколько осталось пропусков после обработки, %

display(pd.DataFrame(round(data.isna().mean()*100, )).style.background_gradient('coolwarm'))
```

```python
# Заполним оставшиеся пустые значения колонки rating значением N/A. Для этого добавим соответствующую категорию в переменную

data['rating'] = data['rating'].cat.add_categories(['N/A'])
data['rating'] = data['rating'].fillna("N/A")
```

```python
# Проверяем
print(data.rating.isna().sum()) #Порядок, пропусков нет
```

```python
# Посмотрим на значения user_score
print(data['user_score'].value_counts())
# В таблице много записей со значением tbd - 2376. Переведем их в NaN и затем тип данных в float. 
#Удалять их не будем пока, т.к. понадобятся для исследования на которые они не влияют
data.loc[(data['user_score'] == 'tbd'), 'user_score'] = 'NaN'

#Проверяем
print("Значений tbd стало: ", data.loc[(data['user_score'] == 'tbd'), 'user_score'].count())
print(f'Пустых значений колонки user_score: ', data['user_score'].isna().sum())
print(data['user_score'].value_counts())

# Меняем тип данных на float
data['user_score'] = data['user_score'].astype('float')

data.info()

```

```python
# Определим корреляцию между оценками юзеров и критиков. Если корреляция сильная, то можно заполнить данные одной таблицы с помощью другой, рассчитав их зависимость
print("Коэф. корреляции: ", data['user_score'].corr(data['critic_score'])) # Коэф корреляции недостаточно большой, чтобы заполнять по их зависимости не вариант
```

## Шаг 5. Проверить гипотезы

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Задайте самостоятельно пороговое значение alpha.

Поясните:

Как вы сформулировали нулевую и альтернативную гипотезы;

Какой критерий применили для проверки гипотез и почему.

**Формулируем нулевую и альтернативную гипотезы**

H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Рейтинг Xbox One == рейтинг PC

H1: Средние пользовательские рейтинги платформ Xbox One и PC различны. Рейтинг Xbox One != рейтинг PC

```python
alpha = 0.05

# Формируем выборки. Берем значения оценок юзеров, фильтруем по платформам. Наны удаляем.
data_after_2012_xone = data_after_2012.loc[data_after_2012['platform']=="XOne", "user_score"].dropna()
data_after_2012_pc = data_after_2012.loc[data_after_2012['platform']=="PC", "user_score"].dropna()

# вызоваем метод для проверки гипотезы. Предполагаем, что дисперсии не равны, выставляем параметр equal_var
results = st.ttest_ind(
    data_after_2012_xone, 
    data_after_2012_pc, equal_var = False)

# вывод значения p-value на экран 
print(results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")

print(data_after_2012_xone.describe())
print(data_after_2012_pc.describe())
```

```python
# Посмотрим на описание всей таблицы
print(data_after_2012.describe())
```

**Формулируем нулевую и альтернативную гипотезы**

H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые. Жанр Action == жанр Sports

H1: Средние пользовательские рейтинги жанров Action и Sports разные. Жанр Action != жанр Sports



```python
# Принимаем альфу за 0.05, так как данных у нас не так много
alpha = 0.05

# Формируем выборки. Берем значения оценок юзеров, фильтруем по жанрам. Наны удаляем.
data_after_2012_action = data_after_2012.loc[data_after_2012['genre']=="Action", "user_score"].dropna()
data_after_2012_sports = data_after_2012.loc[data_after_2012['genre']=="Sports", "user_score"].dropna()

# вызоваем метод для проверки гипотезы. Предполагаем, что дисперсии не равны, выставляем параметр equal_var
results = st.ttest_ind(
    data_after_2012_action, 
    data_after_2012_sports, equal_var = False)

# вывод значения p-value на экран 
print(results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")

print(data_after_2012_action.describe())
print(data_after_2012_sports.describe())
```

Нулевую гипотезу выбрали именно такой, потому что она должна включать знак равенства. H1 выбрали как ей противоположную.

Метод проверки гипотез был выбран о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, так как входные данные м использовали выборки по платформам и жанрам, а затем сравнивали их.
