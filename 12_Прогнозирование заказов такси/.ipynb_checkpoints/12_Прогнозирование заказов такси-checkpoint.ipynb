# Проект "Прогнозирование заказов такси"

## Описание проекта

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики RMSE на тестовой выборке должно быть не больше 48.

### Описание данных

Данные лежат в файле /datasets/taxi.csv. Скачать датасет. 

Количество заказов находится в столбце 'num_orders' (от англ. number of orders, «число заказов»).

## 1. Загрузка и ресемплирование данных

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics as metrics
import lightgbm as lgb
import time
import warnings

from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, cross_val_score, TimeSeriesSplit
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor 
from numpy.random import RandomState
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from catboost import Pool, CatBoostRegressor, cv
from sklearn.preprocessing import OrdinalEncoder

from statsmodels.tsa.seasonal import seasonal_decompose
```

## 3. Обучение моделей

```python
#Для построения фрейма в конце проекта создадим списки
models_list = ['LinearRegression', 'CatBoost', 'LightGBM', 'RandomForestRegressor']

rmse_train_list = []
rmse_test_list = []
```

```python
# Напишем функцию, для создания фичей. Сделаем фичами год, месяц, число, час, день недели
def make_features(data, max_lag, rolling_mean_size):
#    data['year'] = data.index.year 
#    data['month'] = data.index.month
    data_new = data.copy()
    data_new['day'] = data_new.index.day
    data_new['hour'] = data_new.index.hour
#    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1): # Будем также вычисляьть скользящую среднюю
        data_new['lag_{}'.format(lag)] = data_new['num_orders'].shift(lag)

    data_new['rolling_mean'] = data_new['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data_new

data_new = make_features(data, 10, 10)
```

```python
# Делим выборку 1 к 9 по условию проекта. Не мешаем данные

train, test = train_test_split(data_new, shuffle=False, test_size=0.1)
train = train.dropna() # Удаляем значения с NaN
```

```python
features_train = train.drop('num_orders', axis=1)
features_test = test.drop('num_orders', axis=1)
target_train = train.num_orders
target_test = test.num_orders

pred_previous_train = train['num_orders'].shift()
pred_previous_train = pred_previous_train.dropna()

pred_previous_test = test['num_orders'].shift()
pred_previous_test[0] = train['num_orders'][-1]
```

### Модель LinearRegression

```python
model_linreg = LinearRegression()

mse = cross_val_score(model_linreg, features_train, target_train, cv=TimeSeriesSplit(), scoring='neg_mean_squared_error').mean() * -1 

params = {}

model_linreg_cv = RandomizedSearchCV(model_linreg, params, scoring = 'neg_mean_squared_error', cv=TimeSeriesSplit(), n_iter=1, n_jobs=-1, verbose=1)
model_linreg_cv.fit(features_train, target_train)

print('RMSE на обучающей:', mse**0.5)
rmse_train_list.append(mse**0.5)
```

### Модель CatBoost

```python
model_cat = CatBoostRegressor(learning_rate=0.5, random_state=42, 
                              verbose=False, loss_function = "RMSE"
                             )

param_cv = {
            'learning_rate': [0.05, 0.1],
            'n_estimators': range(100, 500, 100),
            }
model_cat_cv = RandomizedSearchCV(model_cat,
                                 param_cv, scoring = 'neg_mean_squared_error', 
                                 cv=TimeSeriesSplit(), n_iter=1, n_jobs=-1, verbose=False
                                )
model_cat_cv.fit(features_train, target_train)
rmse = (model_cat_cv.best_score_ * -1) ** 0.5
print('RMSE на тренировочных данных:', rmse)
rmse_train_list.append(rmse)
```

### Модель LightGBM

```python
model_lgb = LGBMRegressor(random_seed=42, n_jobs=1, objective='regression',num_leaves=31,learning_rate=0.05,n_estimators=10)

param_cv = {
    'learning_rate': [0.05, 0.1],
    'n_estimators': range(100, 500, 100),
}

model_lgb_cv = RandomizedSearchCV(model_lgb, param_cv, scoring = 'neg_mean_squared_error', cv=TimeSeriesSplit(), n_iter=1, n_jobs=-1, verbose=False)
```

```python
model_lgb_cv.fit(features_train, target_train)

print("RMSE на обучающей выборке: ", (model_lgb_cv.best_score_*-1)**0.5, 'при', model_lgb_cv.best_params_)
rmse_train_list.append((model_lgb_cv.best_score_*-1)**0.5)
```

### Модель Случайного леса

```python
param = {'n_estimators': range(1,1000,10), 'max_depth': range(1,31)}
model_rf = RandomForestRegressor(random_state = 42)

model_rf_cv = RandomizedSearchCV(model_rf, param, scoring = 'neg_mean_squared_error', cv=TimeSeriesSplit(), n_iter=1, n_jobs=-1, verbose=1)
model_rf_cv.fit(features_train, target_train)

print("RMSE на обучающей выборке: ", (model_rf_cv.best_score_*-1)**0.5, 'при', model_rf_cv.best_params_)
rmse_train_list.append((model_rf_cv.best_score_*-1)**0.5)
```

**Вывод: Увеличивая параметры max_lag, rolling_mean_size видно, что качество модели растет. Кросс-валидацией обучили модели, rmse записали в словарь.**

## 5. Выбор лучшей модели и финальное тестирование

```python
pd.DataFrame({'RMSE на обучающей выборке':rmse_train_list}, index=models_list)
```

```python
# Посчитаем RMSE для каждой модели CatBoost

predict_cat = model_cat_cv.predict(features_test)
rmse_test = mean_squared_error(target_test, predict_cat) ** 0.5
rmse_test_list.append(rmse_test)
print("RMSE на тестовой выборке модели CatBoost: ", mean_squared_error(target_test, predict_cat, squared=False))
```

**Все модели, кроме линейной регрессии показали сходий RMSE на тренировочных данных. Немного лучше отработала модель CatBoost. Ее и выберем для финального тестирования.**
