# Прогнозирование температуры плавления стали

### Условие задачи

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Вам предстоит построить модель, которая предскажет температуру стали.

### Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- data_arc.csv — данные об электродах;
- data_bulk.csv — данные о подаче сыпучих материалов (объём);
- data_bulk_time.csv — данные о подаче сыпучих материалов (время);
- data_gas.csv — данные о продувке сплава газом;
- data_temp.csv — результаты измерения температуры;
- data_wire.csv — данные о проволочных материалах (объём);
- data_wire_time.csv — данные о проволочных материалах (время).

Во всех файлах столбец key содержит номер партии. В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

### 1.2 Напишем функции, которые будем использовать по ходу выполнения проектной работы

```python
# Напишем функцию, которая покажет общую информацию о датафрейме
def show_dataframe_info(data, name):
    rows, columns = data.shape
    print(f"\033[1m{name}\033[0m")
    print(f'В датафрейме количество стоблцов равно: ', columns)
    print(f'В датафрейме количество строк равно:', rows)
    display(data.head(5))
    display(data.describe())
    data.info()
    print("")
```

```python
# Функция, выводящая распределение данных в столбцах датафрейма

def show_df_plots(df,  title='', x_label='', y_label=''):
    plt.figure(figsize=[12,9])
    for column in df.columns:
        plt.hist(df[column], bins=50, label = column, alpha=0.5)
    plt.legend(loc='upper right')  
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(loc='upper right')
    plt.title(title)
    plt.show()
```

```python
# Функция, выводящая ящик с усами и распределение одной колонки одного датафрейма

def show_subplots(df, column, title_1='', title_2='', x_label_1='', y_label_1='', x_label_2='', y_label_2='', bins=100):
    plt.subplot (1, 2, 1)
    df[column].plot(figsize=(10, 5), bins=bins, fontsize=15, kind='hist', grid=True)
    plt.xlabel(x_label_1, fontsize=15)
    plt.ylabel(y_label_1, fontsize=15)
    plt.title(title_1)

    plt.subplot (1, 2, 2)
    df[column].plot(figsize=(10, 5), kind='box', fontsize=15)
    plt.xlabel(x_label_2, fontsize=15)
    plt.ylabel(y_label_2, fontsize=15)
    plt.subplots_adjust(wspace=0.4, hspace=0)
    plt.title(title_2)
    plt.show()
```

```python
# Функция для переименования столбцов в датафрейме. Пробелы заменяем на _ и переводим в нижний регистр. При необходимости проводим транлитерацию параметром to_eng

def rename_cols(df, to_eng = False):
    counter = 0
    df.columns = [ col.replace(" ", "_") for col in df.columns]
    df.columns= df.columns.str.strip().str.lower()
    if to_eng:
        df.columns = [translit(col, "ru", reversed=True) for col in df.columns]
    return df
```

```python
# Функция для вывода количества всех пропущенных значений из всех датафреймов. На вход подается словарь с датафреймами

def null_value_counts(df_voc):
    for key ,value in df_voc.items():
        display(pd.DataFrame(round(value.isna().mean()*100, 2).sort_values(ascending=False), columns=[key]).style.background_gradient('coolwarm'))
```

```python
# Напишем функцию, которая будет удалять партию целеком, если в ней были замечены выбросы.
# Просто так удалять строки с выбросами мы не можем, так как это будет удаление одной из операций при обработке партии. Нужно удалять партию целеком.
# Ф-ция принимает значение датафрейма и колонки.
def delete_outliers(df, cols):
    q1 = df[cols].quantile(0.25)
    q3 = df[cols].quantile(0.75)   
    iqr = q3 - q1
    bad_keys = df.loc[(df[cols] < (q3 + 1.5 * iqr)) & (df[cols] < (q1 - 1.5 * iqr)), 'key'] 
    return df[~df['key'].isin(bad_keys.values)]
```

```python
# Определяем таргет и признаки
target = data_final['temperatura_last']
features = data_final.drop('temperatura_last', axis=1)
features = data_final[['temperatura_first', 'bulk_1', 'bulk_2', 'bulk_3', 'bulk_4', 'bulk_5',
       'bulk_6', 'bulk_7', 'bulk_8', 'bulk_9', 'bulk_10', 'bulk_11', 'bulk_12',
       'bulk_13', 'bulk_14', 'bulk_15', 'wire_1', 'wire_2', 'wire_3', 'wire_4',
       'wire_5', 'wire_6', 'wire_7', 'wire_8', 'wire_9', 'total_time_per_key',
       "aktivnaja_moschnost'_sum", 'division_power_sum',
       "reaktivnaja_moschnost'_sum", 'total_time_nagreva_dugoj_sum', 'gaz_1',
       ]]

# features = data_final[['temperatura_first', 'bulk_2', 'bulk_4',
#        'bulk_6',  'bulk_11', 'wire_1', 'wire_2', 'total_time_per_key',
#         'gaz_1', #'rabota_sum'
#        ]]

# Делим выборку на тренировочные и тестовые данные

features_train, features_test, target_train, target_test = train_test_split(
   features, target, test_size=TEST_SIZE, random_state=RANDOM_STATE) 

print("Размер тренировочных данных:", features_train.shape)
print("Размер тестовых данных:", features_test.shape)
```

```python
# С помощью модели KNN найдем и удалим аномалии в тренировочной выборке
model = KNN()

model.fit(features_train)
features_train['is_outlier'] =  model.predict(features_train) == 1
outliers_knn = features_train['is_outlier'].sum()
print("Количество аномалий: ", outliers_knn)

outlier_indexes = list(features_train[features_train['is_outlier'] == 1].index)
good_indexes = list(set(features_train.index) - set(outlier_indexes))

features_train = features_train.drop(outlier_indexes)
target_train = target_train.drop(outlier_indexes)
features_train = features_train.drop('is_outlier' , axis=1)
features_train.shape
```

### 2.5 Модель линейной регрессии

```python
# масштабируем признаки для линейной регрессии
scaler = StandardScaler()
scaler.fit(features_train)
features_train_linreg = scaler.transform(features_train)

model_linreg = LinearRegression()
rfecv = RFECV(estimator=LinearRegression())
pipeline = Pipeline([('Feature Selection', rfecv), ('Model', model_linreg)])
cv = RepeatedStratifiedKFold(n_splits=2, n_repeats=2, random_state=RANDOM_STATE)

mae = cross_val_score(pipeline, features_train_linreg, target_train, cv=cv, scoring='neg_mean_absolute_error') 

select = pipeline.fit(features_train_linreg, target_train) # обучаем модель на тренировочной выборке

print('MAE:', mae.mean()*-1)

print('Оптимальное количество признаков : %d' % rfecv.n_features_)

mask = rfecv.get_support()
mask
# features = features
best_features = features.columns[mask]
 
print("Всего признаков: ", features.shape[1])

print("Отобрано признаков: ", best_features.shape[0])
#print("Отобранные признаки", features.columns[mask]) 

rfecv_df = pd.DataFrame(rfecv.ranking_,index=features.columns,columns=['Rank']).sort_values(by='Rank',ascending=True)
display(rfecv_df)

plt.figure(figsize=(12,6))
plt.xlabel('Количество признаков')
plt.ylabel('Оценка по кросс-валидации')
plt.plot(range(1, len(rfecv.ranking_) + 1), rfecv.ranking_)
plt.show()
```

### 2.6 Модель CatBoost

```python
model_cat = CatBoostRegressor(
                          verbose=0, loss_function = "RMSE"
                             )

param_cv = {
            'learning_rate': [0.5, 0.1],
            'n_estimators': range(100, 500, 10)
            }
model_cat_cv = RandomizedSearchCV(model_cat,
                                 param_cv, scoring = 'neg_mean_absolute_error', 
                                 cv=5, n_iter=2, n_jobs=-1, verbose=1, random_state=RANDOM_STATE
                                )
model_cat_cv.fit(features_train, target_train)
mae = (model_cat_cv.best_score_ * -1)
print('MAE на валидационных данных:', mae)

print("Время обучения модели:", model_cat_cv.cv_results_['mean_fit_time'][0])
print("Время предсказаний модели:", model_cat_cv.cv_results_['mean_score_time'][0])
```

- Модель линейной регрессии
- Модель CatBoost
- Модель LightGBM
- Модель случайного леса

### Выбор лучшей модели

Выбор лучшей модели осуществлялся на основе лучшей метрики MAE в GridSearchCV и RandomSearchCV. Скорость обучения модели в расчет не бралась, так как объем данных небольшой и это не представляет проблему. 

Лучшие результаты на кросс-валидации показала модель Catboost и линейная регрессия. . MAE=5.21, 5.23 соответственно на валидационных данных, что вполне удовлетворяет условию заказчика. Ввиду того, что линейная регрессия работала на основе подбора оптимального количества признаков, модель несет определенные риски, т.к. заказчика это может не удовлетворить. Решено было отобрать для тестирования модель CatBoost.

Стоит также отметить, что сыпучие материалы и проволочные добавки добавлялись далеко не равномерно, а значит, часы частые их них никак нельзя исключать из признаков. И наоборот, самыми редкими - можно пренебречь.

### Тестирование модели

Тестирование модели CatBoost показало среднюю абсолютную ошибку MAE=5.55, что практически совпадает с оценкой, которую мы получили в ходе тренировки модели. Отличный результат.

Так же бала протестирована модель линейной регрессии. MAE на тесте сосоавило 5.7, что тоже довольно неплохо.

Поиск наиболее влияющий признаков на предсказания конечной температуры
После тестирования модели были выявлены признаки, которые оказывают наибольшее влияние на результат предсказания модели:
- Начальная температура расплава
- Суммарная активная мощность, затраченная на изготовление одной партии
- Общее время, затраченное на изготовление партии
- Суммарное время нагрева дугой на партию
- Объем добавленной проволоки типа 1
- Объем сыпучих добавок типа 6
- Объем инертного газа
- Синтетический признак division_power_sum
- Объем добавленной проволоки типа 2
- Суммарная реактивная мощность, затраченная на изготовление одной партии
- А также в меньшей степени объемы сыпучих добавок типов 14, 12, 15, 4, 3, 1

## Рекомендации заказчику

1. С учетом выявленных важностей признаков видно, что первичная температура расплава является самым значимым. Сделовательно считаю целесообразным Заказчику предпринять действия к уменьшению теплопотерь, т.к. это непостредственно отразиться на энергозатратах завода.

2. Исследовать новые научные и технологические подходы к уменьшению реактивной мощности, что также снизит энергозатраты.

3. Рассмотреть возможность разработки моделей, удовлетворяющех условиям минимального потребления энегрии и ресурсов с учетом гарантированного попадания в допустимый интервал конечной температуры, что позволит в перспективе выстроить оптимальный производственный процесс.
