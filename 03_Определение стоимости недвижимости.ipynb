```python
#Найдем все значения высоты потолка более 20 метров. Скорее всего ошибка в разряде данных. Поделим на 10. Строки со значенем NaN не трогаем.
data['ceiling_height'].where(
                            (data['ceiling_height'] < 20) | (data['ceiling_height'].isna()),
                            data['ceiling_height']/10,inplace=True)
```

```python
# В столбце с высотой потолков много пропусков (более 9тыс). Заполним медианами пустые значения
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())

# Проверяем
print("Пустых значений в высоте потолка после заполнения медианами: ", data['ceiling_height'].isna().sum())


#Вряд ли в квартирах выота потолков превышает 4 метра. Узнаем сколько таких в базе
print("Выше 4 метров: ", data.loc[data['ceiling_height']>4, 'ceiling_height'].count())

#Всего 53. Выкидываем их
data = data.query('ceiling_height <= 4 | ceiling_height.isna()').reset_index(drop=True)

# Тоже самое сделаем с потолками ниже 2 метров
print("Ниже 2 метров до обрабоки: ", data.loc[data['ceiling_height'] < 2, 'ceiling_height'].count())
data = data.query('ceiling_height > 2').reset_index(drop=True)
#Проверяем
print("Ниже 2 метров после обработки: ", len(data.loc[data['ceiling_height'] < 2]))

# Проверяем
print("Теперь осталось строк с данными: ", len(data))

data['ceiling_height'].plot.box();
```

```python
# Напишем функцию, которая удаляет строки с выбросами. Ф-ция принимает значение датафрейма и колонки.
def delete_outliers(df, cols):
    q1 = df[cols].quantile(0.25)
    q3 = df[cols].quantile(0.75)   
    iqr = q3 - q1
    df = df.loc[(df[cols] < (q3 + 1.5 * iqr)) & (df[cols] > (q1 - 1.5 * iqr))]
    return df
```

```python
# Жилая площадь
# Посмотрим на данные жилой площади и ее распределение
print(data.living_area.describe())

data['living_area'].hist(bins=20).set_xlabel('Распределение жилой площади')
plt.show()
data.boxplot(column='living_area').set_xlabel('Распределение')
plt.show()

# Удалим выбросы
data = delete_outliers(data, 'living_area')

#Проверяем
print(data['living_area'].describe()) # Подозрительно выглядят записи с площадью 2 метра, однако при этом они не являются выбросами. 

#Удалим строки с жилой площадью менее 10 метров. Их 24.
print("Количество записей с жилой площадью менее 10 метров: ", len(data.query('living_area < 10')))
data = data.query('living_area >= 10') # удаляем их

#Посмотрим как распределились теперь значения

data.boxplot(column='living_area').set_xlabel('Распределение жилой площади после обработки');
```

```python
# Площадь кухни

# Посмотрим на данные кухонь и их распределение
print(data.kitchen_area.describe())

data['kitchen_area'].hist(bins=20).set_xlabel('Распределение площадей кухонь до обработки')
plt.show()
data.boxplot(column='kitchen_area').set_xlabel('Распределение площадей кухонь до обработки')
plt.show()


# Определим корреляцию площади кухонь и общей площади
print("Коэф. корреляции: ", data['kitchen_area'].corr(data['total_area'])) # Коэф корреляции 0.55. Можно значит расчитывать площать кухонь отнсительно общей площади

print("Медиана: ", data['kitchen_area'].median())
print(data['kitchen_area'].describe())
data_kitchen = data.loc[:, ('kitchen_area', 'total_area')]
data_kitchen['ratio'] = data_kitchen['kitchen_area'] / data_kitchen['total_area']
print(data_kitchen['ratio'].median())

print("Пустые значения площади кухонь со студиями: ", len(data.query('kitchen_area.isna() == True')))

# Рассчитаем площадь кухонь как общую площадь умноженную на медиану найденного отношения.
# Студии тоже возьмем в расчет, потому как, если поставить нули, то это может повлиять на дальнейшее исследование
data.loc[data['kitchen_area'].isna() == True,'kitchen_area'] = data['total_area'] * data_kitchen['ratio'].median()

# Проверяем
print("Оставшиеся пустые значения кухонь", data['kitchen_area'].isna().sum())

#Проверяем
print(data['kitchen_area'].describe()) # Подозрительно выглядят записи с площадью 2 метра, однако при этом они не являются выбросами. 

#Удалим строки с площадью кухни менее 3 метров и более 25. 
print("Количество записей с площадью кухни менее 3 метров и более 25: ", len(data.query('(kitchen_area < 3) | (kitchen_area > 25)'))) # Таких 361
data = data.query('(kitchen_area >= 3) & (kitchen_area <= 25)') # удаляем их


#Посмотрим как распределились теперь значения
data.boxplot(column='kitchen_area').set_xlabel('Распределение жилой площади после обработки')
plt.show()
data['kitchen_area'].hist().set_xlabel('Распределение площади кухонь')
plt.show()
```
